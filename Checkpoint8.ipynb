# 1. LOAD DATA

```python
import pandas as pd
data=pd.read_csv("CC_GENERAL.csv")
data
```

## Preprocessing Data

```python
print (data.isnull().sum())
```

```python
data['CREDIT_LIMIT'].fillna(data['CREDIT_LIMIT'].mean(),inplace=True)
data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].mean(),inplace=True)
print (data.isnull().sum())
```

```python
data=data.drop("CUST_ID", axis=1)

data
```

# 2. HIERACHICAL CLUSTERING

```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(data) #Applying agglomerative algorithm with 5 clusters, using euclidean distance as a metric
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

# 3. PLOT CLUSTERS

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (data ['BALANCE'] , data ["PURCHASES"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clutering")
ax.set_xlabel("BALANCE")
ax.set_ylabel("PURCHASES")
plt.colorbar(scatter)
```

# 4. DENDOGRAMME

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(data, method="complete"))
```

## Utiliser k-means clustering

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=5, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(data)
```

```python
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
kmeans.predict(data)
print(kmeans.cluster_centers_)
```

# 5. DIFFERENT VALUES OF K

```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km=km.fit(data)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow method for optimal k')
plt.show()
```

La meilleure valeur de k est 8.

# 6. PLOT CLUSTERS

```python
plt.scatter(data["BALANCE"][kmeans.labels_ == 0],          
            data["PURCHASES"][kmeans.labels_ == 0],s=80,c='magenta',label='Prudents')
plt.scatter(data["BALANCE"][kmeans.labels_ == 1],
           data["PURCHASES"][kmeans.labels_ == 1],s=80,c='yellow',label='Imprudents')
plt.scatter(data["BALANCE"][kmeans.labels_ == 2],
           data["PURCHASES"][kmeans.labels_ == 2],s=80,c='green',label='Modérés')
plt.scatter(data["BALANCE"][kmeans.labels_ == 3], 
           data["PURCHASES"][kmeans.labels_ == 3],s=80,c='cyan',label='Standard')
plt.scatter(data["BALANCE"][kmeans.labels_ == 4], 
           data["PURCHASES"][kmeans.labels_ == 4],s=80,c='burlywood',label='épargnants')
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3], s=300, c='red', label = 'Centroides')
plt.title('Clusters of Customers')
plt.xlabel('BALANCE')
plt.ylabel('PURCHASES')
plt.legend()
plt.show()
```

# 7. COMPARE RESULTS

Les deux résultats sont assez similaires, sauf pour les valeurs en haut du graphique.  
En effet pour 'hierarchical clustering' les points qui sont en haut représentent un cluster à eux seuls.  
Tandis que pour 'k-means clustering', ces points sont rattachés à d'autres points un peu plus bas dans un cluster.
